```python
import os
import subprocess
```

```python
nets = ['emails', 'mouse_brain', 'airports']
```

```python
net_dir = "results/airports"

for file in os.listdir(net_dir):
    
    if file == '.DS_Store': continue
    
    # if file is not directory, skip
    if not os.path.isdir(net_dir + "/" + file):
        continue

    for emb_file in os.listdir(net_dir +  "/" + file +"/" ):

        # if there is no result folder, create one
        if not os.path.exists(net_dir + "/" + file +"/results"):
            os.makedirs(net_dir + "/" + file + "/results")

        if emb_file in ['.DS_Store', 'results']: continue
        
        # skip
        command = [
            "julia",
            "./CGE_CLI.jl",
            "-g",
            "airports_reindexed.csv",
            "-e",
            os.path.join( net_dir + "/" + file +"/", emb_file),
            "--result_name",
            net_dir + "/" + file +"/results" +  "/" + emb_file + "_result.csv",
            "-v"
        ]
        output = subprocess.check_output(command, shell=False)
```
