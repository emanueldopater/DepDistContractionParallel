```python
import networkx as nx
from node2vec import Node2Vec
import os
```

```python
# load graph from edgelist
airport_net = nx.read_edgelist('../results/remapped_nets/airports_reindexed.csv', create_using=nx.Graph())
emails_net = nx.read_edgelist('../results/remapped_nets/emails_reindexed.csv', create_using=nx.Graph())
mouse_brain_net = nx.read_edgelist('../results/remapped_nets/mouse_brain_reindexed.csv', create_using=nx.Graph())
```

```python
emb_dims = [4, 8, 16, 32, 64, 128]
iterations = 10
```

```python
for dim in emb_dims:
    for i in range(iterations):

      # check if exists directory and if no create 
      res_dir = 'results/airports/' + 'dim_' + str(dim)
      if not os.path.exists(res_dir):
          os.makedirs(res_dir)

      # Generate walks
      node2vec = Node2Vec(airport_net, dimensions=dim, workers=10)

      # Learn embeddings
      model = node2vec.fit(window=10)

      # Save embeddings for later use
      model.wv.save_word2vec_format(res_dir + '/emb_' + str(dim) + '_' + str(i) + '.emb', write_header=False)
```

```python
# load graph from edgelist
ca_grqc_net = nx.read_edgelist('../results/remapped_nets/ca_grqc_reindexed.csv', create_using=nx.Graph())
ca_hepth_net = nx.read_edgelist('../results/remapped_nets/ca_hepth_reindexed.csv', create_using=nx.Graph())
facebook_combined_net = nx.read_edgelist('../results/remapped_nets/facebook_combined_reindexed.csv', create_using=nx.Graph())
lastfm_asia_reindexed = nx.read_edgelist('../results/remapped_nets/lastfm_asia_reindexed.csv', create_using=nx.Graph())
p2p_gnutella08_reindexed = nx.read_edgelist('../results/remapped_nets/p2p_gnutella08_reindexed.csv', create_using=nx.Graph())
```

```python
nets = [ca_grqc_net, ca_hepth_net, facebook_combined_net, lastfm_asia_reindexed, p2p_gnutella08_reindexed]
names_nets = ['ca_grqc', 'ca_hepth', 'facebook_combined', 'lastfm_asia', 'p2p_gnutella08']
```

```python
for net, name in zip(nets, names_nets):
    for dim in emb_dims:
        for i in range(iterations):

          # check if exists directory and if no create 
          res_dir = 'results/' + name + '/' + 'dim_' + str(dim)
          if not os.path.exists(res_dir):
              os.makedirs(res_dir)

          # Generate walks
          node2vec = Node2Vec(net, dimensions=dim, workers=10)

          # Learn embeddings
          model = node2vec.fit(window=10)

          # Save embeddings for later use
          model.wv.save_word2vec_format(res_dir + '/emb_' + str(dim) + '_' + str(i) + '.emb', write_header=False)
```
